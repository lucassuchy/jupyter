## Matplotlib

Quando for plotar algum grafico, podemos partir dos exemplos que temos no site do matplotlib
https://matplotlib.org/stable/gallery/index.html
---

Começamos importando a biblioteca do matplotlib

```python
import matplotlib.pyplot as plt
```

Para ver os graficos internamente no jupyter usamos

```python
%matplotlib inline
```

Importamos o numpy,
Generamos duas variaveis, uma alimentada com 11 elementos, igualmente espaçados entre 0 e 5.
A outra variavel, é a primeira variavel elevado ao quadrado

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

Verificamos o conteudo as variaveis

```python
x
```

```python
y
```

Agora com os elementos criados, podemos passar eles para o plt e teremos o grafico plotado.

```python
plt.plot(x,y)
```

Para adicionar titulos aos eixos do grafico, usamos as funções xlabel e ylabel
Usamos a função title para definirmos um titulo que vai aparecer no grafico

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

Se quiseremos plotar dois graficos juntos usamos a função subplot que recebe alguns argumentos
que eu vou ver mais adiante

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```
